```python
def plot_hist_artblocks_scripts(df_lengths):
    # Human number formatting
    def human_format(num,pos,dec=0):
        magnitude = 0
        while abs(num) >= 1000:
            magnitude += 1
            num /= 1000.0
        # add more suffixes if you need them
        return f'%.{dec}f%s' % (num, ['', 'K', 'M', 'B', 'T', 'P'][magnitude])
    
    # Create fig
    fig, ax = plt.subplots(figsize=(20,10))
    plt.hist(df_lengths[(df_lengths['Code Length']>0)]["Code Length"],alpha=0.85,color="#483AAA",bins=75)
    plt.title("Distribution of Art Blocks Generative Code Length",fontsize=34)
    plt.vlines(df_lengths[(df_lengths['Code Length']>0)]["Code Length"].median(),0,35,color='black',linestyle='--',label="Median {}".format(human_format(df_lengths[(df_lengths['Code Length']>0)]["Code Length"].median(),[],1)))
    ymin,ymax = ax.get_ylim()
    xmin,xmax = ax.get_xlim()
    plt.ylim(0,ymax)
    plt.xlim(0,xmax)
    plt.xticks(fontsize=28,color='black')
    plt.yticks(fontsize=28,color='black')
    plt.xlabel("Length of Code (Number of Characters)",fontsize=28,color='black')
    plt.ylabel("Number of Projects",fontsize=28,color='black')
    plt.legend(fontsize=28,loc='upper right',facecolor='white')
    formatter = FuncFormatter(human_format)
    ax.xaxis.set_major_formatter(formatter)
```
